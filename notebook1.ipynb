## Ejercicio 1 - Procesamiento de Datos con Pandas:
Pregunta: Dado un conjunto de datos que contiene información de ventas, calcula el total de ventas por categoría de producto.

```python
#Solución:
import pandas as pd
# Crear un DataFrame de ejemplo
data = {'producto': ['A', 'B', 'A', 'C', 'B'],
        'venta': [100, 200, 150, 50, 120]}
df = pd.DataFrame(data)
# Calcular el total de ventas por categoría de producto
total_ventas_por_categoria = df.groupby('producto')['venta'].sum()
print(total_ventas_por_categoria)
```

## Ejercicio 2 - Análisis Exploratorio de Datos:
Pregunta: Dado un conjunto de datos de calificaciones de películas, realiza un análisis exploratorio y muestra un histograma de las calificaciones.

```python
#Solución:
import pandas as pd
import matplotlib.pyplot as plt
# Crear un DataFrame de ejemplo
data = {'pelicula': ['A', 'B', 'C', 'D', 'E'],
        'calificacion': [4, 3, 5, 4, 2]}
df = pd.DataFrame(data)
# Análisis exploratorio: mostrar histograma de calificaciones
plt.hist(df['calificacion'], bins=5, edgecolor='black')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones')
plt.show()
```

## Ejercicio 3 - ETL Básico:
Pregunta: Tienes un archivo CSV con información de ventas. Diseña un pequeño pipeline ETL para cargar los datos en un DataFrame y calcular el promedio de ventas por mes.

```python
#Solución:
import pandas as pd
# Cargar datos desde el archivo CSV
data = pd.read_csv('ventas.csv',sep=';')
```

```python
data
```

```python
# Convertir la columna 'fecha' a tipo datetime
data['fecha'] = pd.to_datetime(data['fecha'])
# Crear una columna adicional para el mes
data['mes'] = data['fecha'].dt.month
# Calcular el promedio de ventas por mes
promedio_ventas_por_mes = round(data.groupby('mes')['venta'].mean(), 2)
print(promedio_ventas_por_mes)
```

## Ejercicio 4 - Modelado Predictivo Simple:
Pregunta: Tienes un conjunto de datos de alturas y pesos de personas. Ajusta un modelo de regresión lineal para predecir el peso basado en la altura.

```python
#Solución:
import pandas as pd
from sklearn.linear_model import LinearRegression
# Crear un DataFrame de ejemplo
data = {'altura': [160, 170, 175, 180, 185],
        'peso': [60, 70, 75, 85, 90]}
df = pd.DataFrame(data)
# Definir variables de entrada (X) y variable objetivo (y)
X = df[['altura']]
y = df['peso']
# Crear y ajustar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# Predecir el peso para una altura específica
altura = [[int(input('Ingrese la altura: '))]]
altura_nueva = altura
peso_predicho = modelo.predict(altura_nueva)
print('Peso predicho:', round(peso_predicho[0],2))
# https://parzibyte.me/blog/2020/12/15/python-imprimir-flotante-dos-decimales/
```

## Ejercicio 5 - Programación:
Pregunta: Escribe una función en Python que tome una lista de números y devuelva la suma de todos los números en la lista.


```python
#Solución:
def suma_lista(lista):
    suma = 0
    for numero in lista:
         suma += numero
    return suma
```

```python
suma_lista([9,2,3,5,10])
```

## Ejercicio 6 - Programación:
Pregunta: Haz una lista con los caracteres de la cadena. Si yn cáracter de la cadena es un espacio o un dígito mayor que 5, elimínalo y no lo inlcuyas en la matriz.

```python
#Solución:
def str_to_list( str ) :
	new_list = []
	for index in range(len(str)):
		if str[index] == ' ':
			continue
		elif str[index].isdigit():
			if int(str[index]) > 5:
				continue
		new_list.append(str[index])

	return new_list
```
